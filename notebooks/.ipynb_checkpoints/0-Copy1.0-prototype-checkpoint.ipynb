```python
df_X = pd.read_csv("../data/raw/train_values.csv")
df_y = pd.read_csv("../data/raw/train_labels.csv")
df_raw = df_X.merge(df_y, on='building_id')
df_raw_test = pd.read_csv("../data/raw/test_values.csv")
```

# Train-Validation Set Creation

```python
X_train, X_val, _, _ = train_test_split(
    df_raw.values,
    df_raw['damage_grade'].values,
    stratify=df_raw['damage_grade'].values,
    test_size=0.33,
    random_state=3
)

df_raw_train = pd.DataFrame(X_train, columns=df_raw.columns)
df_raw_val = pd.DataFrame(X_val, columns=df_raw.columns)
```

```python
print(df_raw_train.shape)
print(df_raw_val.shape)
print(df_raw_test.shape)

plt.figure(figsize=(15, 2.5))
plt.hist([df_raw_train['damage_grade'].values, df_raw_val['damage_grade'].values], density=True)
plt.title("Training and Validation Labels")
plt.show()
```

# Transform Data for ML

```python
df_interim_train = df_raw_train
df_interim_val = df_raw_val
df_interim_test = df_raw_test
```

```python
# Categorical: one hot encoding
df_interim_train = pd.concat(
    [df_interim_train, pd.get_dummies(df_interim_train[cat_feats])],
    axis=1
)
df_interim_val = pd.concat(
    [df_interim_val, pd.get_dummies(df_interim_val[cat_feats])],
    axis=1
)
df_interim_test = pd.concat(
    [df_interim_test, pd.get_dummies(df_interim_test[cat_feats])],
    axis=1
)
```

```python
# Categorical: mean label encoding
for feat in cat_feats:
    mappings = (
        pd.concat(
            [df_interim_train[feat], pd.get_dummies(df_interim_train['damage_grade'], prefix='damage_grade')],
            axis=1
        )
        .groupby(feat).mean()
        .to_dict()
    )
    
    df_interim_train["{}_labelEnc_1".format(feat)] = df_interim_train[feat].map(mappings['damage_grade_1'])
    df_interim_train["{}_labelEnc_2".format(feat)] = df_interim_train[feat].map(mappings['damage_grade_2'])
    df_interim_train["{}_labelEnc_3".format(feat)] = df_interim_train[feat].map(mappings['damage_grade_3'])
    df_interim_val["{}_labelEnc_1".format(feat)] = df_interim_val[feat].map(mappings['damage_grade_1'])
    df_interim_val["{}_labelEnc_2".format(feat)] = df_interim_val[feat].map(mappings['damage_grade_2'])
    df_interim_val["{}_labelEnc_3".format(feat)] = df_interim_val[feat].map(mappings['damage_grade_3'])
    df_interim_test["{}_labelEnc_1".format(feat)] = df_interim_test[feat].map(mappings['damage_grade_1'])
    df_interim_test["{}_labelEnc_2".format(feat)] = df_interim_test[feat].map(mappings['damage_grade_2'])
    df_interim_test["{}_labelEnc_3".format(feat)] = df_interim_test[feat].map(mappings['damage_grade_3'])
```

```python
# Catigorical: drop original columns
df_interim_train = df_interim_train.drop(cat_feats, axis=1)
df_interim_val = df_interim_val.drop(cat_feats, axis=1)
df_interim_test = df_interim_test.drop(cat_feats, axis=1)
```

```python
n_feats = df_interim_train.drop(['damage_grade', 'building_id'], axis=1).shape[1]
n = df_interim_train.shape[0]

print("Num features: {}".format(n_feats))
print("Features to data ratio: {:.3}".format(n_feats / n))
display(df_interim_train.head())
```

# Modeling

```python
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
```

```python
X_train = df_interim_train.drop(['damage_grade', 'building_id'], axis=1).values
X_val = df_interim_val.drop(['damage_grade', 'building_id'], axis=1).values
X_test = df_interim_test.drop(['building_id'], axis=1).values

y_train = df_interim_train['damage_grade'].values
y_val = df_interim_val['damage_grade'].values
```

## XGBoost Multi-Class Classifier

```python
%%time
clf_xgb = xgb.XGBClassifier(
    n_estimators=500,
    objective='multiclass:softmax',
    base_score=0.5
)

clf_xgb.fit(
    X_train,
    y_train,
    eval_set=[(X_train, y_train), (X_val, y_val)],
    eval_metric='merror',
    early_stopping_rounds=10,
    verbose=False
)
```

```python
pred_train = clf_xgb.predict(X_train)
pred_val = clf_xgb.predict(X_val)
pred_test = clf_xgb.predict(X_test)
```

```python
# Evaluate
score_train = f1_score(y_train, pred_train, average='micro')
score_val = f1_score(y_val, pred_val, average='micro')
print("Train f1 score: {}".format(score_train))
print("Val f1 score:   {}".format(score_val))
```

## XGBoost Classifier Aggregator

```python
y_train_c1 = (y_train == 1).astype(int)
y_train_c2 = (y_train == 2).astype(int)
y_train_c3 = (y_train == 3).astype(int)

y_val_c1 = (y_val == 1).astype(int)
y_val_c2 = (y_val == 2).astype(int)
y_val_c3 = (y_val == 3).astype(int)
```

```python
%%time
def fit_clf_xgb(y_train, y_val):
    clf_xgb = xgb.XGBClassifier(
        n_estimators=100,
        objective='binary:logistic',
        base_score=0.5
    )

    clf_xgb.fit(
        X_train,
        y_train,
        eval_set=[(X_train, y_train), (X_val, y_val)],
        eval_metric='aucpr',
        early_stopping_rounds=10,
        verbose=False
    )
    
    return clf_xgb

clf_xgb_c1 = fit_clf_xgb(y_train_c1, y_val_c1)
clf_xgb_c2 = fit_clf_xgb(y_train_c2, y_val_c2)
clf_xgb_c3 = fit_clf_xgb(y_train_c3, y_val_c3)
```

```python
proba_train_c1 = [row[1] for row in clf_xgb_c1.predict_proba(X_train)]
proba_train_c2 = [row[1] for row in clf_xgb_c2.predict_proba(X_train)]
proba_train_c3 = [row[1] for row in clf_xgb_c3.predict_proba(X_train)]

proba_val_c1 = [row[1] for row in clf_xgb_c1.predict_proba(X_val)]
proba_val_c2 = [row[1] for row in clf_xgb_c2.predict_proba(X_val)]
proba_val_c3 = [row[1] for row in clf_xgb_c3.predict_proba(X_val)]

# Select class with highest probability prediction
pred_train = np.argmax(np.array([proba_train_c1, proba_train_c2, proba_train_c3]).T, axis=1) + 1
pred_val = np.argmax(np.array([proba_val_c1, proba_val_c2, proba_val_c3]).T, axis=1) + 1
```

```python
# Evaluate
score_train = f1_score(y_train, pred_train, average='micro')
score_val = f1_score(y_val, pred_val, average='micro')
print("Train: {}".format(score_train))
print("Val:   {}".format(score_val))
```
